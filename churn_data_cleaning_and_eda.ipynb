```python
import pandas as pd

file_path = '../data/E Commerce Dataset.xlsx'

# Load the actual data sheet
df = pd.read_excel(file_path, sheet_name='E Comm')

# Show the first few rows
df.head()
```

# Exploratory Data Analysis (EDA)

```python
# Basic info (Rows and columns)
df.shape 
```

```python
# Column names and data types
df.info()
```

```python
# Summary statistics
df.describe(include='all')
```

```python
# Check null values
df.isnull().sum().sort_values(ascending=False)
```

```python
# Fill numeric columns with reasonable defaults

# If tenure is missing, assume 0 (new user)
df['Tenure'] = df['Tenure'].fillna(0)

# Assume if no order history, use 0
df['OrderAmountHikeFromlastYear'] = df['OrderAmountHikeFromlastYear'].fillna(0)
df['OrderCount'] = df['OrderCount'].fillna(0)
df['CouponUsed'] = df['CouponUsed'].fillna(0)
df['WarehouseToHome'] = df['WarehouseToHome'].fillna(df['WarehouseToHome'].median())
df['HourSpendOnApp'] = df['HourSpendOnApp'].fillna(df['HourSpendOnApp'].median())
df['DaySinceLastOrder'] = df['DaySinceLastOrder'].fillna(df['DaySinceLastOrder'].max())

# If any other nulls left, fill with 0
df = df.fillna(0)

# Check cleanup
df.isnull().sum().sort_values(ascending=False)
```

# EDA with Visual Charts

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set(style='whitegrid')
plt.figure(figsize=(9, 6))

# Churn by Gender
sns.countplot(data=df, x='Gender', hue='Churn')
plt.title("Churn Count by Gender")
plt.show()
```

```python
# Churn by Preferred Login Device
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='PreferredLoginDevice', hue='Churn')
plt.title("Churn Count by Login Device")
plt.xticks(rotation=45)
plt.show()
```

```python
# Churn by Satisfaction Score
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='SatisfactionScore', hue='Churn')
plt.title("Churn by Satisfaction Score")
plt.show()
```

```python
# Order Count vs Churn  Boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Churn', y='OrderCount')
plt.title("Order Count Distribution by Churn")
plt.show()
```

```python
# Step 5: Export cleaned data
output_csv = '../output/cleaned_churn_data.csv'
output_excel = '../output/cleaned_churn_data.xlsx'

# Save as CSV
df.to_csv(output_csv, index=False)

# Save as Excel
df.to_excel(output_excel, index=False)

print(" Cleaned data saved to:")
print("", output_csv)
print("", output_excel)
```
